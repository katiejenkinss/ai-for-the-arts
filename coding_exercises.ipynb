# Working on Algorithms, Code and Pseudocode

If we view algortihms as step-by-step procedures, we can think of everday tasks in a similar manner to help us understand the distinction between algortihms, code and pseudocode.

For example, we can think of washing our hands as an algorithm - 
1. Turn on the tap
2. Wet your hands
3. Apply hand soap/ hand wash
4. Rub hands together, lathering soap and scrubbing in circular motions
5. Rinse hands with water
6. Turn off tap
7. Dry hands 

Or making a bowl of cereal - 
1. Get a bowl
2. Get your chosen cereal 
3. Pour the cereal into the bowl
4. Take milk out of the fridge and pour over the cereal
5. Get a spoon
6. Enjoy your cereal!



### Turning my everday algorithms into pseudocode with the help of CoPilot

After trying to write my own pseudo code using the algorithms above, Microsoft Copilot gave me improvements and this was the response - 

INPUT: water, soap, towel
PRECONDITION: Remove jewelry/watches; ensure water and soap are available

TURN ON tap
WET hands with water
APPLY soap to hands
SET timer to 20 seconds
WHILE timer not expired:
    RUB hands together
    SCRUB between fingers, under nails, and wrists
END WHILE
RINSE hands thoroughly with water
TURN OFF tap
DRY hands completely using towel
POSTCONDITION: Hands are clean and dry
OUTPUT: clean hands



### Turning my pseudo code into python code 

With the help of copilot, I turned my pseudo code into python code, which can be seen below.


```python

import time

def wash_hands():
    # Preconditions
    water_available = True
    soap_available = True
    towel_available = True

    if not water_available or not soap_available:
        print("Cannot wash hands properly. Please ensure water and soap are available.")
        return

    print("Step 1: Turn on tap.")
    print("Step 2: Wet hands with water.")
    print("Step 3: Apply soap to hands.")

    print("Step 4: Rub hands together for 20 seconds.")
    start_time = time.time()
    while time.time() - start_time < 20:
        # Simulate scrubbing details
        print("...scrubbing between fingers, under nails, and wrists...")
        time.sleep(5)  # Pause to simulate time passing

    print("Step 5: Rinse hands thoroughly with water.")
    print("Step 6: Turn off tap.")

    if towel_available:
        print("Step 7: Dry hands completely using towel.")
    else:
        print("No towel available. Air dry your hands.")

    print("✅ Hands are clean and dry!")

# Run the function
wash_hands()
```
