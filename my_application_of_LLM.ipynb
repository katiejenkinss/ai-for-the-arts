# Local Voices AI: Interactive Storytelling for Cultural Inclusion

For my task, I have been asked to present a design for an application that uses a large language model. Here is the design I have created for my application, "Local Voices AI".

##### __What are Large Language Models? (LLMs)__ 

- Large language models (LLMs) are a type of artificial intelligence trained on huge quantities of text data from publications, websites, books, and other sources.  They pick up linguistic patterns and apply them to produce text that is human-like, respond to enquiries, condense data, and tell stories.  In order to generate cohesive and pertinent responses, LLMs anticipate the subsequent word in a sequence by modifying millions or billions of parameters during training.  They are able to engage with users in conversational, natural ways as a result.

Using an LLM, my design, Local Voices AI, highlights under-represented communities by creating interactive, personalised stories about local history, culture, and customs when a user inputs a place or query.  This allows the application to create instructive, interesting, and inclusive experiences while exhibiting how humans can interact with AI to understand culture and heritage.

[LLM Visualisation](https://bbycroft.net/llm)



##### __The user experience__

- User input: User types a location or community (e.g., “Glasgow, London”).
- LLM processing: The AI generates a narrative summarizing historical, cultural, and social information.
- AI output: User receives a short story including perspectives from underrepresented communities.

Follow-up interaction: User can ask questions such as:
“Who were key figures in this community?”
“Are there local festivals today?”

Iterative response: AI adapts narratives based on user queries.


##### __How LLMs Work in this Design__

- Training: LLMs are trained on large, diverse text datasets (books, news, local archives).
- Mechanism: Predicts the next word in a sequence to generate coherent, human-like narratives.

Capabilities used:
- Summarization of historical/cultural information.
- Question-answering for user follow-ups.
- Creative and compassionate text generation for engaging storytelling.


##### __Cultural engagement, inclusion and learning__

Local Voices AI creates interactive, customised stories about local history and culture by utilising the power of large language models.  The AI enables learners to engagingly examine customs and viewpoints from marginalised populations by producing human-like language and answering user enquiries.  By highlighting under-represented voices the design encourages inclusivity and makes cultural knowledge relevant and approachable.  In this sense, immersive learning and significant cultural involvement are directly supported by the LLM's capacity to provide context-aware responses.

##### __Condiderations to be had in relation to data sharing__ 

- Transparency concerns: Many cloud-based LLM services (e.g., ChatGPT) do not clearly explain how user data is stored or used.
- Privacy risk: User inputs such as location or community information could be logged or used for model training without explicit consent.
- Model size issues: Bigger models trained on massive datasets can amplify biases and increase environmental costs - larger is not always better for local storytelling.


Local/smaller models: Initiatives like TinyLlama, Llamafile, and Ollama allow models to run locally on a user’s computer.

Benefits include:
User data remains private and on-device.
Models can be fine-tuned with culturally sensitive, local content.
Offline execution supports communities with limited internet connectivity


##### __Critical reflection__

The possibilities and responsibility of employing LLMs in cultural contexts are highlighted by Local Voices AI.  While technology can foster learning, participation, and inclusiveness, it may also repeat biases, misrepresent communities, or provide deceptive information.  Human control, chosen sources, and proper labelling of AI-generated writing are vital.  Using smaller, local models helps protect privacy and enables culturally relevant storytelling.  Overall, the design demonstrates how knowledge of LLMs can be used morally to improve education and inclusivity while taking social and cultural factors into account.

```python
# Multiple stories for one location
location = "Brixton"
stories = [
    "Story 1: Brixton hosts an annual street festival celebrating music and food.",
    "Story 2: Brixton is home to a rich Afro-Caribbean cultural heritage.",
    "Story 3: Brixton played a key role in community activism during the 1980s."
]

print(f"\nLocal Voices AI: Stories about {location}:")
for story in stories:
    print("-", story)
```
